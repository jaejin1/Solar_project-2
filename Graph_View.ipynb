```python
import pickle
import numpy as np

#모델 불러오기
list_pickle_path = 'model_v1.pkl'
list_unpickle = open(list_pickle_path, 'rb')

clf = pickle.load(list_unpickle)
```

```python
from sklearn import tree
import graphviz

dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("graph")
```

```python
#feature_name = np.array(['temperature', 'Co2', 'inclination'])
feature_name = np.array(['temperature', 'Co2', 'X','Y','Z'])
class_name = np.array(['safe', 'fire', 'X_Fallen','Y_Fallen','Z_Fallen'])
print(feature_name)
print(class_name)
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=feature_name,  
                         class_names=class_name,  
                         filled=True, rounded=True,  
                         special_characters=True)
graph = graphviz.Source(dot_data) 
graph
```
